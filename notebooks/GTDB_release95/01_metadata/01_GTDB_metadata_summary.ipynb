## CheckM

# Selecting the best rep per GTDB species

````r
# filtering out genomes lacking species classifications
df %>%
    filter(gtdb_representative == 't') %>%
    nrow %>% print

df_s = df %>%
    filter(gtdb_representative == 't') %>%
    mutate(SPECIES = gsub('.+;s__', '', gtdb_taxonomy)) %>%
    filter(SPECIES != '') 

df_s %>% nrow %>% print
````

````r
# filtering out UBA genomes
df_s %>% nrow %>% print
df_s = df_s %>%
    filter(!grepl('^UBA[0-9]+$', accession)) 

df_s %>% nrow
````

````r
# filtering out genomes lacking NCBI assembly accession
df_s %>% nrow %>% print
df_s = df_s %>%
    filter(ncbi_genbank_assembly_accession != 'none') 

df_s %>% nrow
````

````r
# ranking based on checkM data
df_s = df_s %>%
    group_by(gtdb_taxonomy) %>%
    mutate(checkm_completeness_rank = min_rank(-checkm_completeness),
           checkm_contamination_rank = min_rank(checkm_contamination),
           summed_rank = min_rank(checkm_completeness_rank + checkm_contamination_rank),
           min_summed_rank = min(summed_rank)) %>%
    ungroup() 

df_s %>% nrow %>% print
df_s %>% head
````

````r
# assessing
df_s %>%
    dplyr::select(gtdb_taxonomy, checkm_completeness, checkm_contamination,
                  checkm_completeness_rank, checkm_contamination_rank, 
                  summed_rank, min_summed_rank) %>%
    arrange(gtdb_taxonomy) %>%
    head(n=20)
````

````r
# Selecting best representative
df_s_f = df_s %>%
    filter(summed_rank == min_summed_rank) %>%
    group_by(gtdb_taxonomy) %>%
    sample_n(1) %>%
    ungroup() %>%
    dplyr::select(-checkm_completeness_rank, -checkm_contamination_rank, 
                  -summed_rank, -min_summed_rank)

df_s_f %>% nrow %>% print
df_s_f %>% head
````

````r
# duplicate organism names?
df_s_f %>%
    group_by(accession, ncbi_organism_name) %>%
    summarize(n=n(), .groups='drop') %>%
    filter(n>1) %>%
    nrow
````

````r
# ncbi_organism_name = ncbi_organism_name
df_s_f = df_s_f %>%
    unite(ncbi_organism_name, accession, ncbi_organism_name, sep='_', remove=FALSE) 

df_s_f %>%
    group_by(ncbi_organism_name) %>%
    summarize(n=n(), .groups='drop') %>%
    filter(n>1) %>%
    nrow %>% print

df_s_f$ncbi_organism_name %>% head(n=10) %>% print
````

## Adding GTDB taxIDs to metadata

````r
# genome metadata
F = '/ebio/abt3_projects2/databases_no-backup/GTDB/release95/Struo/metadata_1per-GTDB-Spec_gte50comp-lt5cont.tsv'
meta = read.delim(F, sep='\t')
meta %>% dim
meta %>% head
````

````r
# taxdump
F = file.path('/ebio/abt3_projects2/databases_no-backup/GTDB/release95/taxdump/',
              'taxID_info.tsv')
taxIDs = read.delim(F, sep='\t') 
taxIDs %>% dim
taxIDs %>% head
````

````r
# table join
meta = meta %>%
    inner_join(taxIDs %>% dplyr::select(taxID, name) %>% rename('gtdb_taxid' = taxID), 
               c('accession'='name')) 

meta %>% dim
meta %>% head(n=3)
````

### Writing table

````r
F = file.path(work_dir, 'metadata_1per-GTDB-Spec_gte50comp-lt5cont_wtaxID.tsv')
meta %>%
    write.table(file=F, sep='\t', quote=FALSE, row.names=FALSE)
cat('File written:', F, '\n')
````

## Alternative assemblies?

````r
# filtering to alternatives
tax = df_dl_todo$gtdb_taxonomy %>% unique
acc = df_dl_todo$accession %>% unique

df_s = df %>%
    filter(gtdb_taxonomy %in% tax) %>%
    filter(!accession %in% acc) %>%
    filter(checkm_completeness >= 50,
           checkm_contamination < 5) 

df_s %>% dim
````

````r
# filtering out UBA genomes
df_s %>% nrow %>% print
df_s = df_s %>%
    filter(!grepl('^UBA[0-9]+$', accession)) 

df_s %>% nrow
````

````r
# filtering out genomes lacking NCBI assembly accession
df_s %>% nrow %>% print
df_s = df_s %>%
    filter(ncbi_genbank_assembly_accession != 'none') 

df_s %>% nrow
````

````r
# ranking based on checkM data
df_s = df_s %>%
    group_by(gtdb_taxonomy) %>%
    mutate(checkm_completeness_rank = min_rank(-checkm_completeness),
           checkm_contamination_rank = min_rank(checkm_contamination),
           summed_rank = min_rank(checkm_completeness_rank + checkm_contamination_rank),
           min_summed_rank = min(summed_rank)) %>%
    ungroup() 

df_s %>% nrow %>% print
df_s %>% head
````

````r
# assessing
df_s %>%
    dplyr::select(gtdb_taxonomy, checkm_completeness, checkm_contamination,
                  checkm_completeness_rank, checkm_contamination_rank, 
                  summed_rank, min_summed_rank) %>%
    arrange(gtdb_taxonomy) %>%
    head(n=20)
````

````r
# Selecting best representative
df_s_f = df_s %>%
    filter(summed_rank == min_summed_rank) %>%
    group_by(gtdb_taxonomy) %>%
    sample_n(1) %>%
    ungroup() %>%
    dplyr::select(-checkm_completeness_rank, -checkm_contamination_rank, 
                  -summed_rank, -min_summed_rank)

df_s_f %>% nrow %>% print
df_s_f %>% head
````

````r
# ncbi_organism_name = ncbi_organism_name
df_s_f = df_s_f %>%
    unite(ncbi_organism_name, accession, ncbi_organism_name, sep='_', remove=FALSE) 

df_s_f %>%
    group_by(ncbi_organism_name) %>%
    summarize(n=n(), .groups='drop') %>%
    filter(n>1) %>%
    nrow %>% print

df_s_f$ncbi_organism_name %>% head(n=10) %>% print
````

````r
# number of representatives per species
df_s_f_s = df_s_f %>%
    group_by(gtdb_taxonomy) %>%
    summarize(n=n(), .groups='drop') 

df_s_f_s$n %>% summary %>% print
````

````r
# checkM stats
df_s_f$checkm_completeness %>% summary %>% print
df_s_f$checkm_contamination %>% summary %>% print
````

````r
# plotting association
p = df_s_f %>%
    ggplot(aes(checkm_completeness, checkm_contamination, color=mimag_high_quality)) +
    geom_point(alpha=0.3) +
    theme_bw()

options(repr.plot.width=5, repr.plot.height=3, repr.plot.res=200)
plot(p)
````
