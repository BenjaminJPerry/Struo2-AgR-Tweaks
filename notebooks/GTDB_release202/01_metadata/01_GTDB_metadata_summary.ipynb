# Goal

* Summarize GTDB metadata (release 202)
* The data will be used to select which genomes will be used for the `Struo(1|2)` pipeline
* GTDB metadata downloaded from https://data.gtdb.ecogenomic.org/releases/release202/202.0/

# Var

````r
# paths
work_dir = '/ebio/abt3_projects/databases_no-backup/GTDB/release202/Struo2/'
bac_metadata_file = '/ebio/abt3_projects/databases_no-backup/GTDB/release202/metadata/bac120_metadata_r202.tsv'
arc_metadata_file = '/ebio/abt3_projects/databases_no-backup/GTDB/release202/metadata/ar122_metadata_r202.tsv'

# params
threads = 4
````

# Load

````r
# reading in metadata
metadata_files = c(bac_metadata_file, arc_metadata_file)

df = metadata_files %>%
    lapply(function(x) fread(x, sep='\t', check.names=TRUE) %>% select.(-ncbi_date)) %>%
    rbindlist(use.names=TRUE)
df
````

# Summarizing

## Representatives

````r
# number of GTDB representatives
df %>%
    filter.(gtdb_representative == 't') %>%
    nrow
````

````r
# number of GTDB representatives per species
df_s = df %>%
    filter.(gtdb_representative == 't') %>%
    summarize.(n=n.(), 
               .by=c(gtdb_taxonomy))
df_s$n %>% summary_x('GTDB reps per species')
````

````r
# assessing these multi-rep per species genomes
df_s %>%
    filter.(n >= 2) %>%
    head(n=30)
````

````r
# number of GTDB representatives per species-taxid
df_s = df %>%
    filter.(gtdb_representative == 't') %>%
    summarize.(n=n.(), 
              .by=c(ncbi_species_taxid))
df_s$n %>% summary_x('GTDB reps per species taxid')
````

````r
# number of GTDB representatives per species-taxid
df_s = df %>%
    filter.(gtdb_representative == 't') %>%
    summarize.(n=n.(), 
               .by=c(ncbi_species_taxid)) %>%
    filter.(ncbi_species_taxid != 'none')
df_s$n %>% summary_x('GTDB reps per species taxid')
````

## CheckM

````r
summary_x(df$checkm_completeness)
summary_x(df$checkm_contamination)
````

````r
# plotting association
p = df %>%
    filter.(checkm_contamination <= 100) %>%
    ggplot(aes(checkm_completeness, checkm_contamination)) +
    geom_point(alpha=0.3) +
    facet_grid(. ~ gtdb_representative) +
    theme_bw()
p.dims(6,3)
plot(p)
````

# Selecting the best rep per GTDB species

````r
# filtering out genomes lacking species classifications
df %>%
    filter.(gtdb_representative == 't') %>%
    nrow %>% print

df_s = df %>%
    filter.(gtdb_representative == 't') %>%
    mutate.(SPECIES = gsub('.+;s__', '', gtdb_taxonomy)) %>%
    filter.(SPECIES != '') 

df_s %>% nrow %>% print
````

````r
# filtering out UBA genomes
df_s %>% nrow %>% print
df_s = df_s %>%
    filter.(!grepl('^UBA[0-9]+$', accession)) 
df_s %>% nrow
````

````r
# filtering out genomes lacking NCBI assembly accession
df_s %>% nrow %>% print
df_s = df_s %>%
    filter.(ncbi_genbank_assembly_accession != 'none') 
df_s %>% nrow
````

````r
# ranking based on checkM data
df_s = df_s %>%
    mutate.(checkm_completeness_rank = min_rank(-checkm_completeness),
            checkm_contamination_rank = min_rank(checkm_contamination),
            .by=c(gtdb_taxonomy)) %>%
    mutate.(summed_rank = min_rank(checkm_completeness_rank + checkm_contamination_rank),
            .by=c(gtdb_taxonomy)) %>%
    mutate.(min_summed_rank = min(summed_rank),
            .by=c(gtdb_taxonomy)) 
df_s 
````

````r
# assessing
df.dims(20,120)
df_s %>%
    select.(gtdb_taxonomy, checkm_completeness, checkm_contamination,
                  checkm_completeness_rank, checkm_contamination_rank, 
                  summed_rank, min_summed_rank) %>%
    arrange.(gtdb_taxonomy) %>%
    head(n=20)
df.dims()
````

````r
# Selecting best representative
df_s_f = df_s %>%
    filter.(summed_rank == min_summed_rank) %>%
    group_by(gtdb_taxonomy) %>%
    sample_n(1) %>%
    ungroup() %>%
    select.(-checkm_completeness_rank, -checkm_contamination_rank, 
            -summed_rank, -min_summed_rank)
df_s_f
````

## Creating unique org. names

* species names are not unique

````r
# duplicate organism names?
df_s_f %>%
    summarize.(n=n.(), 
               .by=c(ncbi_organism_name)) %>%
    filter.(n>1) %>%
    nrow
````

````r
# duplicate organism names?
df_s_f %>%
    summarize.(n=n.(),
              .by=c(accession, ncbi_organism_name)) %>%
    filter.(n > 1) %>%
    nrow
````

````r
# ncbi_organism_name = ncbi_organism_name
df_s_f = df_s_f %>%
    unite.('TMP', accession, ncbi_organism_name, sep='_', remove=FALSE) %>%
    mutate.(ncbi_organism_name = TMP) %>%
    select.(-TMP)

df_s_f %>%
    summarize.(n=n.(), 
               .by=c(ncbi_organism_name)) %>%
    filter.(n>1) %>%
    nrow %>% print
df_s_f$ncbi_organism_name %>% head(n=10) %>% print
````

## Summarizing filtered

````r
# number of representatives per species
df_s_f_s = df_s_f %>%
    summarize.(n=n.(), 
               .by=c(gtdb_taxonomy))
df_s_f_s$n %>% summary_x('reps per species')
````

````r
# checkM stats
summary_x(df_s_f$checkm_completeness)
summary_x(df_s_f$checkm_contamination)
````
